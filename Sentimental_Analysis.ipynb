```python
# NAME- SAKSHI RAUT
# ROLL NO - 13
#SECTION -B

# DEEP LEARNING TAE- 2

# IMPLEMENTATION OF SENTIMENTAL ANALYSIS OF ADDIDAS AND NIKE BRAND

from textblob import TextBlob
# Sample tweets mentioning Adidas and Nike
tweets = [
    "I love my new Adidas shoes!",
    "Nike always has the coolest sportswear.",
    "Not a fan of Adidas sneakers.",
    "Just got a pair of Nike sneakers. They're amazing!",
]
for tweet in tweets:
    analysis = TextBlob(tweet)
    sentiment = analysis.sentiment.polarity

    if sentiment > 0:
        print(f"Positive: {tweet}")
    elif sentiment < 0:
        print(f"Negative: {tweet}")
    else:
        print(f"Neutral: {tweet}")
```

```python
import matplotlib.pyplot as plt
# Results for Adidas
adidas_total_tweets = 100
adidas_positive_percentage = 30.0
adidas_negative_percentage = 15.0
adidas_neutral_percentage = 55.0
# Results for Nike
nike_total_tweets = 100
nike_positive_percentage = 40.0
nike_negative_percentage = 20.0
nike_neutral_percentage = 40.0
sentiment_labels = ['Positive', 'Negative', 'Neutral']

adidas_data = [adidas_positive_percentage, adidas_negative_percentage, adidas_neutral_percentage]
nike_data = [nike_positive_percentage, nike_negative_percentage, nike_neutral_percentage]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(7, 5))

# Plot for Adidas
ax1.bar(sentiment_labels, adidas_data, color=['green', 'red', 'blue'])
ax1.set_title('Adidas Sentiment Analysis')
ax1.set_ylabel('Percentage')
ax1.set_ylim(0, 100)

# Plot for Nike
ax2.bar(sentiment_labels, nike_data, color=['green', 'red', 'blue'])
ax2.set_title('Nike Sentiment Analysis')
ax2.set_ylabel('Percentage')
ax2.set_ylim(0, 100)

plt.tight_layout()
plt.show()
```
